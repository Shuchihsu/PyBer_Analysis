```python
# % means magie, inline - value displays the chart within the cell for a Jupyter Notebook file
%matplotlib inline

# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan","Feb","Mar","April","May","June","July","Aug","Sept","Oct","Nov","Dec"]
#Set the y-axis to a list of floats as the total fare  in US dollars accumulated for each month.
y_axis =[10.02,23.24,39.20,35.42,32.34,27.04,43.82,10.56,11.85,27.90,20.71,20.09]

# Create the plot

plt.plot(x_axis,y_axis)
plt.xlabel("Month")
plt.ylabel("$")
```

```python
# create the plot with ax.plt()
fig,ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig,ax = plt.subplots()
ax.plot(x_axis,y_axis)
#or we can do this:
# fig = plt.figure()
# ax = fig.add_subplot()

```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label = 'Boston')
# Create labels for the x and y axes
plt.xlabel('Date')
plt.ylabel('Fare($)')
# Set the y limit between 0 and 45
plt.ylim(0,45)
# create a title
plt.title('Pyber Fare by Month')
# Add the legend
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker='D', color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes
plt.xlabel('Date')
plt.ylabel('Fare($)')
# Set the y limit between 0 and 45
plt.ylim(0,45)
#plt.xlim('Jan','Oct')
# create a title
plt.title('Pyber Fare by Month')
# Add a grid
plt.grid()
# Add the legend
plt.legend()
```

```python
# Create the plot with ax.plt():# Create the plot and add a label for the legend
fig,ax = plt.subplots()
ax.plot(x_axis,y_axis,color = 'green', marker ="D", linewidth=2,label='Boston')


# Create labels for the x and y axes
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
# Set the y limit between 0 and 45
ax.set_ylim(0,45)
# create a title
ax.set_title('Pyber Fare by Month')
# Add the legend
ax.legend()
# Add a grid
ax.grid()
```

```python
# Create the plot.
plt.bar(x_axis,y_axis, color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes
plt.xlabel('Date')
plt.ylabel('Fare($)')
# Set the y limit between 0 and 45
plt.ylim(0,45)
# create a title
plt.title('Pyber Fare by Month')
# Add the legend
plt.legend()
```

```python
# Create the plot.
plt.barh(x_axis,y_axis,color="magenta",label="Boston")
plt.gca().invert_yaxis()
# create a title
plt.title('Pyber Fare by Month')
plt.xlabel('Fare($)')
plt.ylabel('Date')
# Add the legend
plt.legend()
```

```python
# Create the plot with ax.plt()
fig,ax = plt.subplots()
ax.bar(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig,ax = plt.subplots()
ax.barh(x_axis,y_axis, color = 'cyan', label='Chicago')
#invert the y axis
ax.invert_yaxis()
# lables
ax.set_ylabel('Date')
ax.set_xlabel('Fare(s)')
#title
ax.set_title('PyBae Fare by Month')
```

```python
# scatter chart
plt.scatter(y_axis,x_axis,s=y_axis, c='red', label='Chicago')
#labels
plt.xlabel('Fare($)')
plt.ylabel('Date')
#set limit for x axis
plt.xlim(0,50)
# invert y axis 
plt.gca().invert_yaxis()
# Add legend
plt.legend()
```

```python
# scatter chart(make the bubbles bigger )
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis,y_axis,s=y_axis_larger)
```

```python
# scatter chart(make the bubbles bigger )

plt.scatter(x_axis,y_axis,s=[i*3 for i in y_axis])
```

```python
# scatter chart with object_oriented interface
fig,ax= plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# scatter chart with object_oriented interface
fig,ax= plt.subplots()
#make the bubbles as big as the fare measurement
ax.scatter( y_axis,x_axis,alpha =0.2, color ='skyblue',s=[i*5 for i in y_axis],edgecolors = "r",marker='o',label='Boston')
# invert x and y
ax.invert_yaxis()
ax.set_xlim(0,50)

ax.legend()
```

```python
# pit chart
explode = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(10,8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.pie(y_axis,labels = x_axis, explode=explode, autopct ="%.1f%%" ,colors = colors)
plt.show()
```

```python
# Pie chart with Object-Oriented interface
explode = (0,0,.2,0,0,0,0.2,0,0,0,0,0)
fix, ax = plt.subplots(figsize=(8,8))

ax.pie(y_axis, labels = x_axis, shadow = True, counterclock =False,explode =explode, startangle=95)

```

```python
import numpy as np
plt.barh(x_axis,y_axis)
# Using ticks
plt.xticks(np.arange(0,51,step=5.0))
plt.gca().invert_yaxis()
```

```python
# Using ticks and object-oriented
fig,ax =plt.subplots()
ax.barh(x_axis,y_axis)
ax.set_xticks(np.arange(0,51,step=5))
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
